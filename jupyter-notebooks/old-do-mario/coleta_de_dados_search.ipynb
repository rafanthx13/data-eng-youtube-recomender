### Processamento dos dados brutos

```python
for query in queries:
    for page in range(1,2):
        # Lendo arquivos html
         with open("video.html".format(query=query, page=page), 'r+') as inp:
#         with open("./dados_brutos/{query}_{page}.html".format(query=query, page=page), 'r+') as inp:
            page_html = inp.read()
            # Processando a página html
            parsed = bs(page_html, "html.parser")
            ancors = parsed.find_all('a', attrs={"id": "video-title"})
            # Buscando todas as tags "a"
            for e in ancors:
                link = e['href']
                title = e['title']
                with open("parsed_videos.json", "a+") as output:
                    data = {"link": link, "title": title, "query": query}
                    output.write("{}\n".format(json.dumps(data)))
```

```python
ydl = youtube_dl.YoutubeDL({"ignoreerrors": True, 'verbose':False, "date": "2020-08-10"})
```

```python
resultados = []
for query in queries:
    r = ydl.extract_info("ytsearch1100:{}".format(query), download=False, extra_info={"date": "2020-08-10"})
    for entry in r['entries']:
        if entry is not None:
            entry['query'] = query
    resultados += r['entries']
```

```python
resultados = [e for e in resultados if e is not None]
len(resultados)
```

```python
today = datetime.datetime.now()
two_days_from_now = (today + datetime.timedelta(2)).strftime("%Y-%m-%d")
```

```python
df = pd.DataFrame(resultados)
df['upload_date'] = pd.to_datetime(df['upload_date'])
df['tempo_desde_pub'] = (pd.to_datetime(two_days_from_now) - df['upload_date']) / np.timedelta64(1,'D')
df.head()
```
